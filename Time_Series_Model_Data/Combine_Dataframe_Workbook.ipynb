# Combine Dataframe Workbook
Workbook to combine the results datframe and the weight prediction dataframe into one dataframe to be exported for Tableau.

```python
# Import Libraries to load dataframe
import pandas as pd
```

```python
# Load the result dataframe
df = pd.read_csv("../Data/combined_pond_result.csv", delimiter=",", low_memory=False)
```

```python
# Display the dataframe
df.head()
```

```python
# Load the weight prediction dataframe
df1=pd.read_csv("../Weight_prediction_Models/Per_minute_Weight_diff_pred_Pond2.csv", delimiter=",", low_memory=False)
df1.head()
```

```python
#  Add the weight difference prediction to the result dataframe
df['Weight_diff_Pred']=df1['Weight_diff_Pred']
```

```python
df.head()
```

```python
# Save the dataframe to csv file
df.to_csv("./Predictions/combined_pond_result.csv", index=False)
```
